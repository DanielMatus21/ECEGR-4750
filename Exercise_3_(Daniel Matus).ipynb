```python
# -*- coding: utf-8 -*-
"""
regression example
Chirag Shah, University of Washington
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import io

# Load the dataset.
df = pd.read_csv(io.BytesIO(uploaded['Boston.csv']))
print(df)
print("correlation coefficient = ", np.corrcoef(df.nox, df.dis) [0,1])

# I tested each variable to check for correlation coefficients, and dis had the strongest correlation coefficient with nox

y = df.nox # dependent variable
X = df.dis # independent variable
X = sm.add_constant(X) # adding the y intercept

lr_model = sm.OLS(y,X).fit() # regression w/ ordinary least squares

print(lr_model.summary())

X_prime = np.linspace(X.dis.min(), X.dis.max(), 100) # Guido took this directly from MATLAB.

X_prime = sm.add_constant(X_prime)

y_hat = lr_model.predict(X_prime)

plt.scatter(df.dis, df.nox)

plt.plot(X_prime[:,1], y_hat, "red", alpha = 0.9)

plt.xlabel("Dis")
plt.ylabel("Nox")
```

```python
# -*- coding: utf-8 -*-
"""
regression example
Chirag Shah, University of Washington
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import io

# Load the dataset.
df = pd.read_csv(io.BytesIO(uploaded['Boston.csv']))
# print(df)
print("correlation coefficient = ", np.corrcoef(df.medv, df.lstat) [0,1])

# I tested each variable to check for correlation coefficients, and lstat had the strongest correlation coefficient with medv

y = df.medv # dependent variable
X = df.lstat # independent variable
X = sm.add_constant(X) # adding the y intercept

lr_model = sm.OLS(y,X).fit() # regression w/ ordinary least squares

print(lr_model.summary())

X_prime = np.linspace(X.lstat.min(), X.lstat.max(), 100) # Guido took this directly from MATLAB.

X_prime = sm.add_constant(X_prime)

y_hat = lr_model.predict(X_prime)

plt.scatter(df.lstat, df.medv)

plt.plot(X_prime[:,1], y_hat, "red", alpha = 0.9)

plt.xlabel("Lstat")
plt.ylabel("Medv")
```
